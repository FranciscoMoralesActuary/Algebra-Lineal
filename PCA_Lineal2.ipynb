```python
#Funcion para PCA
from sklearn.decomposition import PCA
#Funcion para Estadalizar Columnas
from sklearn.preprocessing import StandardScaler
#Libreria para graficas en 2 dimensiones
import matplotlib . pyplot as plt
#Libreria para calculos con visualizar datos
import seaborn as sns
#Libreria para calculos con matrices de datos
import pandas as pd
#Libreria para calculos con numericos
import numpy as np
#Libreria para graficas en 3 dimensiones
import plotly.express as px
```

```python
#Introducimos nuestros datos
datos = {
"Estadistica": [10 , 9, 10, 8 , 9,7,10,5,10,8,6,10,10,9,10,9,9,7,7,6,9,7,8,8,6,8,9,9,9,10] ,
"Probabilidad": [10 , 9, 9 , 8 , 10,6,10,5,10,8,6,10,10,9,10,9,9,7,7,6,7,7,8,6,6,9,9,8,9,10] ,
"Ingles": [9 , 9, 10 , 6 , 6,10,10,5,9,8,6,5,10,9,8,9,9,7,7,6,9,7,8,7,5,9,9,9,10,10] ,
"Algebra": [8 , 8, 8 , 6 , 8,5,10,5,9,8,6,10,9,9,9,10,9,7,7,6,8,6,9,8,5,10,10,7,10,10] ,
"Demografia": [10 , 10, 10 , 10 , 10,10,10,5,10,8,6,10,10,9,10,10,9,7,7,6,8,6,10,8,5,8,10,8,9,10] ,
"Alumno": ["Francisco", "Diego", "Denilson", "Juan", "Tienda","Jesus","David","Yahir","Arturo","Hector",
           "Eduardo","Claudio","Paquito","Erick","Jessica","Rugerio","Luis","Ramon","Martin","Armenta","Ana",
           "Perla","Gabo","Esau","Adan","Apolinar","Marelyn","Oswaldo","Fernanda","Lucio"]
 }
```

```python
#Los volvemos un DataFrame para poder operarlos
df = pd.DataFrame(datos)
df.head(5)
```

```python
#Elegimos solo las columnas Numericas
X = df [[ "Estadistica", "Probabilidad","Ingles","Algebra","Demografia"]]
X.head(5)
```

```python
# Estandarizar columnas
ss = StandardScaler()
df_estandarizado = ss.fit_transform(X)
df_estandarizado
```

```python
#Lo volvemos un Data Frame
datos = pd.DataFrame(df_estandarizado)
datos.head(5)
```

```python
# PCA 2 componentes
pca2 = PCA(n_components=2, random_state=42)
pca_2 = pca2.fit_transform(datos)
pca_2
```

```python
#Lo volvemos un data frame
dataframe_2 = pd.DataFrame({'PCA1': pca_2[:, 0], 'PCA2': pca_2[:, 1]})
dataframe_2.head()
```

```python
# Varianza total explicada
pca2.explained_variance_ratio_
```

```python
# Suma de la varianza
pca2.explained_variance_ratio_.sum()
```

```python
# Crear grafico de la varianza explicada
sns.barplot(x = ['PCA1', 'PCA2'], y = pca2.explained_variance_ratio_)
```

```python
#Creamos el grafico de 2 PCA
sns.scatterplot(x = 'PCA1', y = 'PCA2', data = dataframe_2)
```

```python
# PCA 3 Componentes
pca3 = PCA(n_components=3, random_state=42)
pca_3 = pca3.fit_transform(datos)
pca_3[:3]
```

```python
#Lo volvemos un DataFrame
dataframe_3 = pd.DataFrame({'PCA1': pca_3[:, 0], 'PCA2': pca_3[:, 1] , 'PCA3': pca_3[:, 2]})
dataframe_3.head()
```

```python
# Varianza total explicada
pca3.explained_variance_ratio_
```

```python
# Suma de la varianza
pca3.explained_variance_ratio_.sum()
```

```python
# Crear grafico de la varianza explicada
sns.barplot(x = ['PCA1', 'PCA2', 'PCA3'], y = pca3.explained_variance_ratio_)
```

```python
#Creamos el grafico de  3 PCA
fig = px.scatter_3d(dataframe_3, x='PCA1', y='PCA2', z='PCA3',
              )
fig.show()
```
